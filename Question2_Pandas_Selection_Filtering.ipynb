
#Create a pandas DataFrame using the employee data



```python
import pandas as pd
# create DataFrame:
data={
    "EmployeeID":[201,202,203,204,205,206,207,208,209,210],
    "Name":['Sarah','Michael','Jennifer','David','Lisa','James','Emma','Robert','Sophia','William'],
    "Department":['Enginnering','Sales','Marketing','Enginnering','HR','Sales','Enginnering','Marketing','Sales','HR'],
    "Salary":[85000,62000,71000,92000,58000,68000,78000,75000,65000,60000],
    "Performance Score":[4,3,5,5,4,4,3,4,5,3],
    "Years of Service":[5,3,7,8,4,6,2,5,4,3]
}
employees_data=pd.DataFrame(data)
employees_data
```

#Select and display only the Name, Department, and Salary columns from the employee DataFrame.

```python
employees_data[['Name','Department','Salary']]
```

#Use iloc to retrieve and display the employee located at row position 5 (zero-based indexing)

```python
employees_data.iloc[5]
```

#Use loc to retrieve and display the employee whose EmployeeID is 208

```python
employees_data.set_index('EmployeeID').loc[208]
```

#Filter the DataFrame to display only employees who work in the Engineering department

```python
employees_data[employees_data['Department']=='Enginnering']
```

#Filter the DataFrame to show employees with a salary greater than 70,000

```python
employees_data[employees_data['Salary']>70000]
```

Filter employees from the Sales department who have a performance score of 4 or higher

```python
employees_data[
    (employees_data['Department']=='Sales')&
    (employees_data['Performance Score']>=4)
]
```

#Sort the DataFrame by salary in descending order and display the top three highest-paid employees

```python
employees_data.sort_values('Salary',ascending=False).head(3)
```

#Sort the DataFrame by department in ascending order and by salary in descending order within each department

```python
employees_data.sort_values(['Department','Salary'],ascending=[True,False])
```
